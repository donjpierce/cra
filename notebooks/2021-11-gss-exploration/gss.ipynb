```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sas7bdat import SAS7BDAT
import pandas as pd

with SAS7BDAT('gss2021.sas7bdat') as file:
    gss_df = file.to_data_frame()

df = pd.read_stata('gss2021.dta')
```

```python
def educ_sum(data, col):
    return data.groupby([col]).agg({'EDUC': 'sum'}) / data.groupby([col]).agg({'ID': 'count'}).rename({'ID': 'EDUC'}, axis=1)
```

```python
skip_cols = ['YEAR', 'ID', 'EDUC']
for col in gss_df.columns:
    if col in skip_cols or gss_df[col].isna().all():
        continue
    print(col)
    print(educ_sum(gss_df, col))
    print()
```

```python
def plot_educ_dist(data, col):
    return educ_sum(data, col).hist(by=data[col])
```
