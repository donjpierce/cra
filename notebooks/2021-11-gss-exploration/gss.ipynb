```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sas7bdat import SAS7BDAT
import pandas as pd

with SAS7BDAT('gss2021.sas7bdat') as file:
    gss_df = file.to_data_frame()

df = pd.read_stata('gss2021.dta')
```

```python
def educ_sum(data, col):
    return data.groupby([col]).agg({'EDUC': 'sum'}) / data.groupby([col]).agg({'ID': 'count'}).rename({'ID': 'EDUC'}, axis=1)
```

```python
def plot_educ_dist(data, col):
    return educ_sum(data, col).plot(figsize=(10, 10))
```

## Further Analysis

```python

skip_cols = ['YEAR', 'ID', 'EDUC', 'WTSSNRPS', 'WSSPS']
diffs = {}
for col in gss_df.columns:
    if col in skip_cols or gss_df[col].isna().all():
        continue
    educ = educ_sum(gss_df, col)
    max, min = educ.max(), educ.min()
    diffs[col] = max - min
```

```python
t = pd.DataFrame.from_dict(diffs).T.sort_values('EDUC', ascending=False)
```
